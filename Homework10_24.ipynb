```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

car_data_raw = pd.read_csv("https://cdn.c18l.org/vehicles_lab.csv")
```

```python
car_data_raw.head()
```

I chose year and odometer since they were already in a numerical form. I chose type, condition, and posting date because they are the categories people searching Craigslist consider when searching for a car.

```python
data = car_data_raw.loc[:, ['price', 'year', 'type', 'condition', 'odometer', 'posting_date']]

data.head()
```

```python
temp1 = pd.get_dummies(data['condition'], dummy_na=True).astype(int)
temp2 = pd.get_dummies(data['type'], dummy_na=True).astype(int)
```

```python
new_data = pd.concat([data, temp1, temp2], axis=1)
new_data.head()
```

```python
new_data.describe()
```

I used get_dummies to make the condition and type of car into numerical values. I am not sure if/why this is necessary (if it impacts the model I plan to make later), but since we did it in class, I chose to do it in the homework as well.

In order to make the data a little less varried, I restricted the make year to newer than 1975, the price to under 30 thousand dollars, and restrict the odometer to be between 0 and 200 thousand miles.

```python
new_data = new_data.loc[new_data['price'] <= 30000.0]
new_data = new_data.loc[new_data['year'] > 1975]
new_data = new_data.loc[new_data['odometer'] > 0]
new_data = new_data.loc[new_data['odometer'] <= 200000]
new_data.describe()
```

I wanted to make a new column that is the age of the car when initially posed on Craigslist and a column for miles per year. However, to make the mile_per_year column not be infinite, I am restricting car_age to be more than 0.

```python
new_data['posting_year'] = new_data['posting_date'].str.extract(r'^([0-4][0-4][0-4][0-4])')
new_data['car_age'] = new_data['posting_year'].astype(int) - new_data['year']
new_data = new_data.loc[new_data['car_age'] > 0]
new_data.describe()
```

```python
new_data['miles_per_year'] = new_data['odometer'] / new_data['car_age']
new_data.head()
```

```python
cut_labels_4 = ['1', '2', '3']
cut_bins = [0, 10000, 20000, 30000]
new_data['price_bin'] = pd.cut(new_data['price'], bins=cut_bins, labels=cut_labels_4)
```

```python
new_data = new_data.dropna()
```

```python
new_data.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    new_data,
    train_size=0.8,
    random_state=42)

model = LinearRegression().fit(
    X=train_data.loc[:, [
        'odometer', 'excellent', 'fair', 'good',
        'like new', 'new', 'salvage', 'SUV', 'bus',
        'convertible', 'coupe', 'hatchback', 'mini-van',
        'offroad', 'other', 'pickup', 'sedan', 'truck',
        'van', 'wagon', 'car_age', 'miles_per_year']],
    y=train_data['price_bin']
)
```

```python
model
```

```python
model.score(
    X=test_data.loc[:, [
        'odometer', 'excellent', 'fair', 'good',
        'like new', 'new', 'salvage', 'SUV', 'bus',
        'convertible', 'coupe', 'hatchback', 'mini-van',
        'offroad', 'other', 'pickup', 'sedan', 'truck',
        'van', 'wagon', 'car_age', 'miles_per_year']],
    y=test_data['price_bin']
)
```
