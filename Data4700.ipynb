```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
```

```python
parcels = pd.read_csv('/content/sample_data/Parcels.csv', low_memory = False)
parcels.head(1)
#parcel_number
```

```python
parcels1 = parcels[["parcel_number",
                    "taxpayer_city",
                    "property_class",
                    "property_class_desc",
                    "tax_status",
                    "total_square_footage",
                    "total_acreage",
                    "frontage",
                    "depth",
                    "nez",
                    "num_bldgs",
                    "year_built",
                    "sale_price",
                    "sale_date",
                    "assessed_value_previous",
                    "taxable_value_previous"
                    ]]
parcels2 = parcels1[parcels1["property_class"].isin([401,402])]
parcels2.head(10)
```

```python
parcels2["taxpayer_city"] = np.where(parcels2["taxpayer_city"] == "DETROIT", 1, 0)
```

```python
parcels2['sale_year'] = [x[:4] for x in parcels2['sale_date']]
```
